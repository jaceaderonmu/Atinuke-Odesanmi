```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('main.xlsx')
```

 ## Trendline Chart: GDP (current US$) and GNI (current US$)

```python
gdp = df[df["Indicator Name"] == "GDP (current US$)"]
gni = df[df["Indicator Name"] == "GNI (current US$)"]

# Create a trendline chart
plt.figure(figsize=(10, 6))

plt.plot(gdp.columns[4:], gdp.iloc[0, 4:], label="GDP (current US$)", color='lightgreen', marker='o')
plt.plot(gni.columns[4:], gni.iloc[0, 4:], label="GNI (current US$)", color='darkgreen', marker='o')

plt.title("Trendline Chart: GDP (current US$) and GNI (current US$)")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Column Chart showing Rural population, Urban Population and Population, total by year

```python
rural_population = df[df["Indicator Name"] == "Rural population"]
urban_population = df[df["Indicator Name"] == "Urban population"]
population_total = df[df["Indicator Name"] == "Population, total"]

# Create a column chart
plt.figure(figsize=(10, 6))

bar_width = 0.25
years = rural_population.columns[4:]

plt.bar(np.arange(len(years)), rural_population.iloc[0, 4:], width=bar_width, label="Rural population", color='limegreen')
plt.bar(np.arange(len(years)) + bar_width, urban_population.iloc[0, 4:], width=bar_width, label="Urban Population", color='forestgreen')
plt.bar(np.arange(len(years)) + 2*bar_width, population_total.iloc[0, 4:], width=bar_width, label="Population, total", color='lightgreen')

plt.title("Column Chart showing Rural population, Urban Population and Population, total by year")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.xticks(np.arange(len(years)) + bar_width, years, rotation=45)

plt.tight_layout()
plt.show()
```

## Scatterplot showing the relationship between Import of Goods (Current US$) and  Export of Goods (Current US$)

```python
selected_years = [str(year) for year in range(2001, 2021)]
imports_of_goods = df[df["Indicator Name"] == "Imports of goods and services (current US$)"][selected_years]
export_of_goods = df[df["Indicator Name"] == "Exports of goods and services (current US$)"][selected_years]

# Create a scatter plot
plt.figure(figsize=(10, 6))

green_colors = plt.cm.get_cmap('Greens', len(selected_years))

for i, year in enumerate(selected_years):
    plt.scatter(imports_of_goods[year], export_of_goods[year], label=year, color=green_colors(i))


plt.title("Scatter Plot: Imports of goods and services (current US$) vs Exports of goods and services (current US$)")
plt.xlabel("Imports of goods and services (current US$)")
plt.ylabel("Exports of goods and services (current US$)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Barchart showing Trade (% of GDP), Gross domestic savings (% of GDP) and Agriculture, forestry, and fishing, value added (% of GDP) for year 2011

```python
selected_year = "2011"
indicators_to_plot = ["Trade (% of GDP)", "Gross domestic savings (% of GDP)", "Agriculture, forestry, and fishing, value added (% of GDP)"]
selected_data = df[df["Indicator Name"].isin(indicators_to_plot)][["Indicator Name", selected_year]]

# Create a bar chart with different shades of green
plt.figure(figsize=(10, 6))

# Generate a list of shades of green
green_colors = plt.cm.get_cmap('Greens', len(indicators_to_plot))

for i, indicator in enumerate(indicators_to_plot):
    plt.bar(indicator, selected_data[selected_data["Indicator Name"] == indicator][selected_year].values[0], color=green_colors(i))

plt.title(f"Bar Chart: Indicators for {selected_year}")
plt.xlabel("Indicators")
plt.ylabel("Value")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Pie chart showinig Fertility rate, total (births per woman) by year

```python
fertility_rate = df[df["Indicator Name"] == "Fertility rate, total (births per woman)"].iloc[:, 4:]


plt.figure(figsize=(12, 6))

green_colors = plt.cm.get_cmap('Greens', len(fertility_rate.columns))

for i, year in enumerate(fertility_rate.columns):
    plt.pie(fertility_rate[year], labels=fertility_rate.index, autopct='%1.1f%%', colors=[green_colors(i)], startangle=90)
    plt.title(f"Pie Chart: Fertility rate, total (births per woman) by {year}")
    plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
    plt.show()
```
