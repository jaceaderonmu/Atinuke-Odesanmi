```python
# Visualization of dataset
# Categorical Variables
# Bar plots
# making a function to plot categorical data according to target

import math

def plot_categorical_to_target(telco_df, categorical_values, target):
    number_of_columns = 2
    number_of_rows = math.ceil(len(categorical_values)/2)

    fig = plt.figure(figsize = (12, 5*number_of_rows))

    for index, column in enumerate(categorical_values, 1):
        ax = fig.add_subplot(number_of_rows, number_of_columns, index)
        ax = sns.countplot(x = column, data = telco_df, hue = target, palette="Blues")
        ax.set_title(column)
    return plt.show()
```

```python
customer_services = ["PhoneService", "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]
plot_categorical_to_target(telco_df, customer_services, "Churn")
```

```python
customer_account_cat = ["Contract", "PaperlessBilling", "PaymentMethod"]
plot_categorical_to_target(telco_df, customer_account_cat, "Churn")
```

```python
# Generating Numerical variables
# Histogram plots shown below

def histogram_plots(telco_df, numerical_values, target):
    number_of_columns = 2
    number_of_rows = math.ceil(len(numerical_values)/2)

    fig = plt.figure(figsize = (12, 5*number_of_rows))

    for index, column in enumerate(numerical_values, 1):
        ax = fig.add_subplot(number_of_rows, number_of_columns, index)
        ax = sns.kdeplot(telco_df[column][telco_df[target]=="Yes"], fill = True)
        ax = sns.kdeplot(telco_df[column][telco_df[target]=="No"], fill = True)
        ax.set_title(column)
        ax.legend(["Churn", "No Churn"], loc='upper right')
        plt.savefig("numerical_variables.png", dpi=300)
    return plt.show() 
```

```python
# Creating a histogram of the Tenure column

plt.hist(telco_df['tenure'], bins=20, color='purple')
plt.xlabel('Tenure(months)')
plt.ylabel('Frequency')
plt.show()
```

```python
customer_account_num = ["tenure", "MonthlyCharges", "TotalCharges"]
histogram_plots(telco_df, customer_account_num, "Churn")
```

```python
# Outliers checking with Box plots

def outlier_check_boxplot(telco_df, numerical_values):
    number_of_columns = 2
    number_of_rows = math.ceil(len(numerical_values)/2)

    fig = plt.figure(figsize = (12, 5*number_of_rows))

    for index, column in enumerate(numerical_values, 1):
        ax = fig.add_subplot(number_of_rows, number_of_columns, index)
        ax = sns.boxplot(x = column, data = telco_df, palette="Blues")
        ax.set_title(column)
    plt.savefig("outliers_check.png", dpi=300)
    return plt.show()
```
